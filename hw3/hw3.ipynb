## Part a)
### (i)

```python
class LMS:

    def __init__(self, v, y, z, w, learning_rate):
        self.v = v
        self.y = y
        self.z = z
        self.w = w
        self.learning_rate = learning_rate

    def fit_3D(self):
        N = self.v.shape[0]
        M = self.v.shape[1]
        mse = np.zeros((N, M))
        for i in range(N):
            for j in range(M):
                y_hat = np.inner(self.w[j], self.v[i, j])
                mse[i, j] = (self.y[i, j]-y_hat)**2

                self.w[j] += self.learning_rate*(self.z[i, j]-y_hat)*self.v[i, j]

        return self.w, mse

    def fit_2D(self):
        N = self.v.shape[0]
        mse = np.zeros((N,))
        w_vals = []
        for i in range(N):
            y_hat = np.inner(self.w, self.v[i])
            mse[i] = (self.y[i]-y_hat)**2

            w_vals.append(self.w+self.learning_rate*(self.z[i]-y_hat)*self.v[i])
            self.w += self.learning_rate*(self.z[i]-y_hat)*self.v[i]

        return self.w, mse, w_vals
```

### (ii)
Below are the learning curves and the averaged coefficient graphs for each SNR with both the given learning rates.

```python
w0_3 = np.zeros((matched_3_v.shape[1], matched_3_v.shape[2]))
learning_rate_3_1 = 0.05
learning_rate_3_2 = 0.15
lms_3_1 = LMS(matched_3_v, matched_3_y, matched_3_z, w0_3, learning_rate_3_1)
lms_3_2 = LMS(matched_3_v, matched_3_y, matched_3_z, w0_3, learning_rate_3_2)

w_3_1, mse_3_1 = lms_3_1.fit_3D()
w_3_2, mse_3_2 = lms_3_2.fit_3D()

averaged_mse_3_1 = np.average(mse_3_1, axis=1)
averaged_mse_3_2 = np.average(mse_3_2, axis=1)

averaged_w_3_1 = np.average(w_3_1, axis=1)
averaged_w_3_2 = np.average(w_3_2, axis=1)

n_range = np.arange(0, matched_3_v.shape[0])

plt.figure()
plt.plot(n_range, averaged_mse_3_1)
plt.plot(n_range, averaged_mse_3_2)
plt.xlabel('Sequence Number')
plt.ylabel('MSE')
plt.legend(['Learning Rate = 0.05', 'Learning Rate = 0.15'])
plt.title('3 dB Learning Curve')
plt.show()

_range = np.arange(0, matched_3_v.shape[1])
plt.figure()
plt.plot(_range, averaged_w_3_1)
plt.plot(_range, averaged_w_3_2)
plt.xlabel('Sequence Number')
plt.ylabel('Averaged Coefficients')
plt.legend(['Learning Rate = 0.05', 'Learning Rate = 0.15'])
plt.title('3 dB Coefficient Plot')
plt.show()
```

## Part c)

First in this part, I used LMS and plotted the resulting MSE values over n for several different non-diverging learning rates. These learning rates can be seen below.

```python
learning_rates_mm = [0.005, 0.01, 0.025, 0.05, 0.1, 0.125, 0.15]
w0_mm = np.zeros((len(learning_rates_mm), mismatched_v.shape[1], mismatched_v.shape[2]))
mse_vals = []

for i, lr in enumerate(learning_rates_mm):
    lms_mm = LMS(mismatched_v, mismatched_y, mismatched_y, w0_mm[i], lr)
    w_mm, mse_mm = lms_mm.fit_3D()
    mse_vals.append(mse_mm)

average_mse_mm_vals = []
for m in mse_vals:
    average_mse_mm_vals.append(np.average(m, axis=1))

n_range = np.arange(0, mismatched_v.shape[0])

for avg_mse, lr in zip(average_mse_mm_vals, learning_rates_mm):
    plt.figure()
    plt.plot(n_range, avg_mse)
    plt.xlabel('Sequence Number')
    plt.ylabel('MSE')
    plt.title(f'Mismatched Learning Curves: lr={lr}')

plt.show()
```

Below I computed the LLSE over all n using the LLSE method.

```python
R_vs = []
r_vys = []
r_ys = []
for mm_v, mm_y in zip(mismatched_v, mismatched_y):
    R_vs.append((1/mm_v.shape[0])*np.matmul(mm_v.T, mm_v))
    r_vys.append((1/mm_v.shape[0])*np.matmul(mm_v.T, mm_y))
    r_ys.append((1/mm_y.shape[0])*np.matmul(mm_y.T, mm_y))

llse_vals = []
for r_y, R_v, r_vy in zip(r_ys, R_vs, r_vys):
    llse_vals.append(r_y**2-np.linalg.multi_dot([r_vy.T, R_v, r_vy]))

n_range = np.arange(0, mismatched_v.shape[0])

plt.plot(n_range, llse_vals)
plt.show()
```

```python
lms_mm_vals = []
for a in average_mse_mm_vals:
    lms_mm_vals.append(a[-1])

llse = llse_vals[-1]

print(f'We get that the minimum MSE value for all the learning rates using LMS is: {min(lms_mm_vals)}'
        f' and the MSE value using LLSE is: {llse}. Thus, we can see that LLSE clearly produces a lower estimate than LMS.')
```
